**import** **package**

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets,linear_model
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split   
```

**load** **dataset**

```python
diabetes=datasets.load_diabetes()
x=diabetes.data[:,np.newaxis,2]
y=diabetes.target;
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
```

**fit** **model**

```python
regr=linear_model.LinearRegression() #Linear regression object
regr.fit(x_train,y_train) #Use training sets to train the model
```

**predict test data label**

```python
diabetes_y_pred=regr.predict(x_test) #Make predictions
print(mean_squared_error(y_test,diabetes_y_pred))
sum( np.power(abs(y_test-diabetes_y_pred),2))/len(y_test)
r2_score(y_test,diabetes_y_pred) #Variance score
```

**plot**

```python
plt.scatter(x_test,y_test,color ='black')
plt.plot(x_test,diabetes_y_pred,color='pink',linewidth=3)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
m=regr.coef_
b=regr.intercept_
y_p=np.multiply(x_train,m)+b
y_p
print('Slope:' ,regr.coef_)
print('Intercept:', regr.intercept_)
plt.scatter(x_train,y_train,color ='black')
plt.plot(x_train,y_p,color='pink',linewidth=3)
y_pred=regr.predict(x_train) #Make predictions
plt.plot(x_train,y_pred,color='red',linewidth=3)
plt.xticks(())
plt.yticks(())
plt.show()
```
