```python
import pandas as pd

#creat from a csv
df=pd.read_csv('ai4i2020.csv')
df.head() #return top 5 rows
df.tail() #return bottom 5 rows
#creat from a dictionary
tempdic={'col1':[1,2,3],'col2':[4,5,6],'col3':[7,8,9]}
dictdf=pd.DataFrame.from_dict(tempdic)
```

```python
#show column and data type
df.columns
df.dtypes
#summary statistics
df.describe()
df.describe(include='object')
```

```python
#filtering columns
df.Type
df[['Product ID','Type']]
df.Type.unique()
```

```python
#filtering on rows
df[df['Type']=='M']
df[df['Air temperature [K]']==298.1]
df[(df['Air temperature [K]']==298.1) & (df['Type']=='M')]
```

```python
#indexing with iloc
df.iloc[5] # return 5-th row
df.iloc[5,1]# return index 1 of cloumn from row 5
df.iloc[5,-1]
df.iloc[1:5] 
```

```python
#indexing with loc
state=df.copy()
state.set_index('Type',inplace=True)
state.head()
state.loc['M'] #indexing with non integer
```

```python
#dropping rows
df.isnull().sum()
df.dropna(inplace=True)
#droping column
df.drop('Type',axis=1)
#creating clculated columns
df['new col']=df['Air temperature [K]']+df['Process temperature [K]']
#updating an entire col
df['new col']=150
#updating single value 
df.iloc[1,-1]=100
#condition based updating using apply
df.loc[0:200,'Type']='M'
df.loc[200:df.size,'Type']='N'
df['binary type']=df['Type'].apply(lambda x: 1 if x=='M' else 0)
df[df['Type']=='M'].tail()

# output csv
df.to_csv('output.csv')
```
